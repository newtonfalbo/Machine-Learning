```python
import matplotlib.pyplot as plt
import unittest

def calcular_metricas(VP, VN, FP, FN):
    acuracia = (VP + VN) / (VP + VN + FP + FN)
    precisao = VP / (VP + FP) if (VP + FP) != 0 else 0
    recall = VP / (VP + FN) if (VP + FN) != 0 else 0
    especificidade = VN / (VN + FP) if (VN + FP) != 0 else 0
    f1_score = (2 * precisao * recall) / (precisao + recall) if (precisao + recall) != 0 else 0

    return {
        "Acurácia": acuracia,
        "Precisão": precisao,
        "Recall": recall,
        "Especificidade": especificidade,
        "F1-Score": f1_score
    }

def plotar_metricas(metricas):
    nomes = list(metricas.keys())
    valores = list(metricas.values())

    plt.figure(figsize=(8, 5))
    plt.bar(nomes, valores, color=['blue', 'orange', 'green', 'red', 'purple'])
    plt.xlabel("Métricas")
    plt.ylabel("Valor")
    plt.title("Comparação das Métricas de Avaliação")
    plt.ylim(0, 1)
    plt.show()

# Exemplo de uso com valores arbitrários
VP, VN, FP, FN = 50, 40, 10, 5
metricas = calcular_metricas(VP, VN, FP, FN)
plotar_metricas(metricas)

# Testes automatizados
class TestMetricas(unittest.TestCase):
    def test_metricas(self):
        metricas = calcular_metricas(50, 40, 10, 5)
        self.assertAlmostEqual(metricas["Acurácia"], 0.90, places=2)
        self.assertAlmostEqual(metricas["Precisão"], 0.83, places=2)
        self.assertAlmostEqual(metricas["Recall"], 0.91, places=2)
        self.assertAlmostEqual(metricas["Especificidade"], 0.80, places=2)
        self.assertAlmostEqual(metricas["F1-Score"], 0.87, places=2)

if __name__ == "__main__":
    unittest.main()
```
